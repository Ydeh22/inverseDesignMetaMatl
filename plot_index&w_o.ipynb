```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['1', '2', '3', '4', '5'])
y1 = np.array([4.0709, 4.2533, 4.1307, 4.2493, 4.1771])
y2 = np.array([3.4228, 2.7806, 3.1105, 3.1934, 3.1743])
y3 = np.array([3.3363, 2.8969, 3.1470, 3.1349, 3.2243])

plt.figure(figsize=(10,5))
plt.plot(x, y1, ':', marker='.', label="index list")
plt.plot(x, y2, '-.', marker='.', label="edge list")
plt.plot(x, y3, marker='.', label="Adj matrix")
plt.xlabel('Layers')
plt.ylabel('Loss (RMSE)')
plt.legend()
plt.show()
```

```python
xxx, yy = ['1 layer', '2 layers', '3 layers', '4 layers', '5 layers'], [3.4228, 2.7806, 3.1105, 3.1934, 3.1743]
```

```python
import matplotlib.pyplot as plt
import numpy as np

#x = np.linspace(-2,8, num=301)
#y = np.sinc((x-2.21)*3)


fig, ax = plt.subplots()
ax.plot(xx, yy, marker='o')

xx = np.array(xx)
yy = np.array(yy)

def annot_min(x,y, ax=None):
    xmin = xx[np.argmin(yy)]
    print(xmin)
    ymin = yy.min()
    text= "MSE Loss={:.3f}".format(ymin)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmin, ymin), xytext=(0.94,0.96), **kw)

annot_min(xx,yy)


ax.set_ylim(2.5, 4)
plt.show()
```
